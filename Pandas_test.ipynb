# 🐼 Introducción a Pandas

**Pandas** es la librería de código abierto más popular de Python para el **análisis y la manipulación de datos**. Es una herramienta fundamental en el flujo de trabajo de cualquier Científico de Datos, Analista o Ingeniero de Software, ya que actúa como un puente entre los datos en bruto y la obtención de conclusiones de negocio.

---

### 🚀 ¿Para qué sirve principalmente?

* **Estructuración de datos:** Permite cargar datos desde múltiples fuentes (archivos CSV, Excel, bases de datos SQL o JSON) y transformarlos en tablas bidimensionales llamadas **DataFrames** (muy similares a una hoja de cálculo o tabla de base de datos).
* **Limpieza de información:** Facilita la detección y el manejo de datos faltantes (`NaN`), eliminación de duplicados, corrección de tipos de datos y filtrado de valores atípicos.
* **Transformación y Agrupación:** Permite realizar operaciones complejas de agrupamiento (estilo *Tablas Dinámicas* de Excel) en una sola línea de código mediante funciones como `.groupby()`.
* **Análisis Estadístico Rápido:** Ofrece funciones integradas para calcular medias, medianas, desviaciones estándar, correlaciones y resúmenes estadísticos completos con un solo comando.

---

### 📊 Estructuras de Datos Clave

Pandas se basa principalmente en dos estructuras:

1. **Series:** Una estructura unidimensional (imagínala como una sola columna de una tabla).
2. **DataFrame:** Una estructura bidimensional con filas y columnas (la tabla completa).

> 💡 **En resumen:** Pandas te permite automatizar procesos de Excel, manejar millones de filas de forma eficiente y preparar tus datos antes de pasarlos a modelos de Inteligencia Artificial o gráficos interactivos.

```python
import pandas as pd

# 1. Crear un conjunto de datos de prueba (Diccionario)
datos = {
    'Empleado': ['Ana', 'Luis', 'Carlos', 'Ana', 'Luis', 'Sofía', 'Carlos', 'Sofía'],
    'Departamento': ['Norte', 'Sur', 'Norte', 'Norte', 'Sur', 'Este', 'Sur', 'Este'],
    'Ventas': [2500, 4000, 1500, 3100, 2200, 5000, 1800, 4200],
    'Mes': ['Enero', 'Enero', 'Enero', 'Febrero', 'Febrero', 'Enero', 'Febrero', 'Febrero']
}

# 2. Convertir el diccionario en un DataFrame de Pandas
df = pd.DataFrame(datos)

print("--- DataFrame Original ---")
print(df)
print("\n" + "="*40 + "\n")

# 3. Filtrar datos: Mostrar solo las ventas mayores a 3000
ventas_altas = df[df['Ventas'] > 3000]
print("--- Ventas mayores a 3000 ---")
print(ventas_altas)
print("\n" + "="*40 + "\n")

# 4. Agrupar datos: ¿Cuánto vendió cada empleado en total?
ventas_por_empleado = df.groupby('Empleado')['Ventas'].sum().reset_index()
print("--- Total de ventas por empleado ---")
print(ventas_por_empleado)
print("\n" + "="*40 + "\n")

# 5. Estadísticas rápidas de las columnas numéricas
print("--- Resumen estadístico de las ventas ---")
print(df['Ventas'].describe())
```
