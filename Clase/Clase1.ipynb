```python
# una variable es un espacio en la memoria que se utiliza para almacenar un valor. En Python, las variables se crean automáticamente cuando se les asigna un valor. Por ejemplo, si escribimos x = 5, estamos creando una variable llamada x y asignándole el valor 5. Las variables pueden almacenar diferentes tipos de datos, como enteros, flotantes, cadenas de texto, listas, diccionarios, entre otros.
nombre = 'Sebastian'
Edad = 20
estatura = 1.65
Hombre = True

print('Nombre: ', nombre)
print('Edad: ', Edad)
print('Estatura: ', estatura)
print('Hombre: ', Hombre)

print(f'Nombre: {nombre}', f'Edad: {Edad}', f'Estatura: {estatura}', f'Hombre: {Hombre}') 

# no se puede usar simbolos numeros y palabras reservadas para crear variables. Por ejemplo, no se puede crear una variable llamada 1nombre, ya que comienza con un número. Tampoco se puede crear una variable llamada if, ya que if es una palabra reservada en Python.
# pero si se puede usar si hay caracteres normales, como letras, números y guiones bajos. Por ejemplo, se puede crear una variable llamada nombre_1, ya que comienza con una letra y contiene un guion bajo. También se puede crear una variable llamada nombre1, ya que comienza con una letra y contiene un número.
# tener cuidad con las minusculas y mayusculas, ya que Python es sensible a ellas. Por ejemplo, si creamos una variable llamada nombre y otra llamada Nombre, Python las considerará como dos variables diferentes. Por lo tanto, es importante ser consistente con el uso de mayúsculas y minúsculas al nombrar variables.
```
