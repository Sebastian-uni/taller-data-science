```python
# tuplas () son inmutables, no se pueden modificar, agregar o eliminar elementos
tupla1 = (1, 2, 3, 4, 5)
print(tupla1[0]) # 1
# listas [] son mutables, se pueden modificar, agregar o eliminar elementos
lista1 = [1, 2, 3, 4, 5]
print(lista1[0]) # 1
# diccionarios {} son mutables, se pueden modificar, agregar o eliminar elementos
diccionario1 = {"nombre": "Sebastian", "apellido": "Gonzalez", "edad": 30}
print(diccionario1["nombre"]) # Sebastian
# conjuntos {} son mutables, se pueden modificar, agregar o eliminar elementos
conjunto1 = {1, 2, 3, 4, 5}
print(conjunto1) # {1, 2, 3, 4, 5}
# slicing en tuplas, listas y cadenas de texto
# [inicio:fin:step]
tupla2 = tupla1[0:3] # (1, 2, 3)  
print(tupla2) # (1, 2, 3)
#tupla de tuplas
tupla3 = ((1, 2), (3, 4), (5, 6))
print(tupla3[0][0]) # 1
```

```python
# Append, extend, insert, remove, pop, clear, index, count, sort, reverse
mi_lista = [1, 2, 3, 4, 5]
mi_lista.append(6) # agrega un elemento al final de la lista
print(mi_lista) # [1, 2, 3, 4, 5, 6]
# una lista de listas
lista_de_listas = [[1, 2, 3], [4, 5, 6], [7, 8, 9]]
#sclicing en listas de listas
print(lista_de_listas[0][0]) # 1
# append en listas
mi_lista = [1, 2, 3, 4, 5]
mi_lista.append(6) # agrega un elemento al final de la lista
# extend en listas
mi_lista.extend([7, 8, 9]) # agrega varios elementos al final de
#insert en listas
mi_lista.insert(0, 0) # agrega un elemento en la posicion 0

```

```python
#diccionario 
mi_diccionario = {"nombre": "Sebastian", "apellido": "Gonzalez", "edad": 30}
print(mi_diccionario["nombre"]) # Sebastian
# Se busca los valores de un diccionario por su clave
#metodos
# get() devuelve el valor de la clave especificada
# keys() devuelve una lista de todas las claves del diccionario
# values() devuelve una lista de todos los valores del diccionario
print(mi_diccionario.get("nombre")) # Sebastian
print(list(mi_diccionario.keys())) # ['nombre', 'apellido', 'edad']
print(list(mi_diccionario.values())) # ['Sebastian', 'Gonzalez', 30]
# items() devuelve una lista de tuplas (clave, valor)
print(list(mi_diccionario.items())) # [('nombre', 'Sebastian'), ('apellido', 'Gonzalez'), ('edad', 30)]
# POP() elimina un elemento del diccionario y lo devuelve
print(mi_diccionario.pop("edad")) # 30
print(mi_diccionario) # {'nombre': 'Sebastian', 'apellido': 'Gonzalez'}
```

```python
# Estructuras condicionales
# if, elif, else, if -else, if - elif - else, if anidado, if ternario
a = 10
if a < 20:
    print("a es menor que 20")
else:
    print("a no es menor que 20")
# Condicionales anidados
a = 10
if a < 20:
    if a < 10:
        print("a es menor que 10")
    else:
        print("a es mayor o igual que 10")

# ingreso por teclado
a = int(input("Ingrese un numero 1: "))
b = int(input("Ingrese otro numero 2: "))
c = int(input("Ingrese otro numero 3: "))

if a > b and a > c:
    print("a es mayor que b y c")
elif b > a and b > c:
    print("b es mayor que a y c")
else:
    print("c es mayor que a y b")

```

```python
#comparadores
# ==, !=, >, <, >=, <=
#operadores logicos
# and, or, not
# Realizar un programa que al recibir una califcaion, le deje un mensade de:
# si tuvopor debajo de 3.0 resultado es deficiente
# si obtuvo valor entre 3.0 a 3.5 el resultado es aceptable
# si obtuvo valor entre 3.5 a 4.0 el resultado es bueno
# si obtuvo valor entre 4.0 a 4.5 el resultado es muy bueno
# si obtuvo valor entre 4.5 a 5.0 el resultado es excelente o sobre
nota = float(input("Ingrese su calificacion: "))
if nota < 3.0:
    print("Deficiente")
elif nota >= 3.0 and nota < 3.5:
    print("Aceptable")
elif nota >= 3.5 and nota < 4.0:
    print("Bueno")
elif nota >= 4.0 and nota < 4.5:
    print("Muy bueno")
else:
    print("Excelente")
```

```python
#bucles
# while, for, break, continue, pass
#Ejemplo de bucle while
contraseña = "1234"
clave = input("Ingrese la contraseña: ")
while clave != contraseña:
    print("Contraseña incorrecta")
    clave = input("Ingrese la contraseña: ")
print("Contraseña correcta")
#break, continue, pass
```

```python
# for
frutas = ["manzana", "banana", "cereza", "durazno", "kiwi", "mango", "naranja", "pera", "sandia", "uva"]
 
for fruta in frutas:
    print(fruta)
```

```python
for i in range(10):
    print(frutas[i])
```

```python
#Recorre un diccionario con un bucle for
mi_diccionario = {"nombre": "Sebastian", "apellido": "Gonzalez", "edad": 30}
for clave, valor in mi_diccionario.items():
    print(clave, valor)

for index, dato in enumerate(frutas):
    print(index, dato)
```

```python
# funciones
# una funcion es un bloque de codigo que se puede reutilizar, se define con la palabra reservada def, seguida del nombre de la funcion y parentesis, dentro de los parentesis se pueden definir parametros, y el bloque de codigo se define con sangria.
def saludar(nombre):
    print("Hola " + nombre)

saludar("Sebastian") # Hola Sebastian

# python tiene funciones predefinidas, como print(), input(), len(), type(), range(), etc.
# pero tambien se pueden crear funciones personalizadas, como la funcion saludar() que se definio anteriormente.
def sumar(a, b):
    return a + b

sumar(10, 20) # 30

def restar(a, b):
    return a - b

restar(10, 20) # -10

#Función Cuadratica
```
