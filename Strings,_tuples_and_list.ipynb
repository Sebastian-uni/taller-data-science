Consideremos la variable d y usemos el segmentado para imprimir los tres primeros elementos:

```python
# Write your code below and press Shift+Enter to execute

d = "ABCDEFG"
print(d[0:3])
# d[0:3] devuelve los primeros tres caracteres de la cadena 'd', que son "ABC". El índice 0 indica el primer carácter y el índice 3 indica el cuarto carácter, pero no se incluye en el resultado.
```

Utilice un valor de paso de 2 para imprimir cada segundo carácter de la cadena e:

```python
# Write your code below and press Shift+Enter to execute

e = 'clocrkr1e1c1t'
print(e[::2])
# e[::2] devuelve los caracteres de la cadena 'e' en posiciones pares, es decir, cada segundo carácter comenzando desde el índice 0. El resultado es "correct".
```

Convierte la variable f a mayúsculas:

```python
# Write your code below and press Shift+Enter to execute

f = "You are wrong"
print(f.upper())
# Se utiliza el método upper() para convertir todos los caracteres de la cadena 'f' a mayúsculas. El resultado es "YOU ARE WRONG".
```

Convierta la variable f2 a minúsculas:

```python
# Write your code below and press Shift+Enter to execute
f2="YOU ARE RIGHT"
print(f2.lower())
# Se utiliza el método lower() para convertir todos los caracteres de la cadena 'f2' a minúsculas. El resultado es "you are right".
```

Consideremos la variable g y encontremos el primer índice de la subcadena snow:

```python
# Write your code below and press Shift+Enter to execute

g = "Mary had a little lamb Little lamb, little lamb Mary had a little lamb \
Its fleece was white as snow And everywhere that Mary went Mary went, Mary went \
Everywhere that Mary went The lamb was sure to go"
print(g.find("snow"))
# El método find() se utiliza para buscar la primera aparición de la subcadena "snow"
```

En la variable g, reemplace la subcadena Mary por Bob:

```python
# Write your code below and press Shift+Enter to execute
print(g.replace("Mary", "Bob"))
# El método replace() se utiliza para reemplazar todas las apariciones de la subcadena "Mary" con la subcadena "Bob". El resultado es una nueva cadena donde todas las ocurrencias de "Mary" han sido reemplazadas por "Bob".
```

Considere la siguiente tupla:

```python
# sample tuple

genres_tuple = ("pop", "rock", "soul", "hard rock", "soft rock", \
                "R&B", "progressive rock", "disco")
genres_tuple
```

Acceda al elemento, con respecto al índice 3:

```python
# Write your code below and press Shift+Enter to execute
print(genres_tuple[3])
# Los indices empiezan desde 0, por lo que el índice 3 corresponde al cuarto elemento de la tupla, que es "hard rock".
```

Utilice el segmentado para obtener los índices 3, 4 y 5:

```python
# Write your code below and press Shift+Enter to execute
print(genres_tuple[3:6])
# [3:6] va desde el índice 3 (inclusive) hasta el 6 (exclusivo), obteniendo los elementos en las posiciones 3, 4 y 5.
```

Encuentra los dos primeros elementos de la tupla genres_tuple:

```python
# Write your code below and press Shift+Enter to execute
print(genres_tuple[0:2])
#[0:2] toma desde el índice 0 hasta el 2 (exclusivo), retornando los primeros dos elementos.
```

Encuentra el primer índice de "disco":

```python
# Write your code below and press Shift+Enter to execute
print(genres_tuple.index("disco"))
# .index() en tuplas funciona como .find() en strings: retorna la posición del elemento. "disco" está en la posición 7 (el último).
```

Generar una lista ordenada a partir de la tupla C_tuple=(-5, 1, -3):

```python
# Write your code below and press Shift+Enter to execut
C_tuple = (-5, 1, -3)
C_list = sorted(C_tuple)
print(C_list)
# sorted() convierte la tupla a una lista y la ordena de menor a mayor. El resultado es una lista (corchetes), no una tupla.
```

Crea una lista a_list, con los siguientes elementos 1, hello, [1,2,3] y True.

```python
# Write your code below and press Shift+Enter to execut
a_list = [1, "hello", [1, 2, 3], True]
print(a_list)
# Las listas en Python pueden contener elementos de distintos tipos: enteros, strings, otras listas, booleanos, etc. Todo va entre corchetes [] separado por comas.
```

Encuentra el valor almacenado en el índice 1 de a_list.

```python
# Write your code below and press Shift+Enter to execut
print(a_list[1])
# El índice 0 es 1 (entero), el índice 1 es "hello" (string). Siempre recuerda que los índices empiezan en 0.
```

Recuperar los elementos almacenados en los índices 1, 2 y 3 de a_list.

```python
# Write your code below and press Shift+Enter to execut
print(a_list[1:4])
# [1:4] toma desde el índice 1 hasta el 4 (exclusivo), es decir, los elementos en las posiciones 1, 2 y 3.
```

Concatena las siguientes listas A = [1, 'a'] y B = [2, 1, 'd']:
