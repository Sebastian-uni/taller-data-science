## Parte 1: Creación de la Base de Datos e Importación

### Paso 1.1: Crear la base de datos `ejercicios`
Se usó el motor de bases de datos MySQL con el cliente gráfico de MySQL Workbench y ejecutar la siguiente sentencia SQL:

```python
CREATE DATABASE ejercicios;
```

### Paso 1.2: Crear la tabla `fuel_consumption` de forma manual
A continuación, crearemos la estructura de la tabla con los tipos de datos correctos correspondientes a las columnas del archivo [FuelConsumptionCo2.csv](./FuelConsumptionCo2.csv).

```python
-- Seleccionar la base de datos recién creada
USE ejercicios;

CREATE TABLE fuel_consumption (
    MODELYEAR INT,                         -- Año de fabricación (ej. 2014)
    MAKE VARCHAR(50),                      -- Fabricante / Marca (ej. ACURA)
    MODEL VARCHAR(100),                    -- Modelo del vehículo (ej. ILX)
    VEHICLECLASS VARCHAR(50),              -- Clase de vehículo (ej. COMPACT)
    ENGINESIZE DECIMAL(3,1),               -- Tamaño del motor en litros (ej. 2.0)
    CYLINDERS INT,                         -- Número de cilindros (ej. 4)
    TRANSMISSION VARCHAR(10),              -- Tipo de transmisión (ej. AS5, M6)
    FUELTYPE VARCHAR(5),                   -- Tipo de combustible (ej. Z, D, X)
    FUELCONSUMPTION_CITY DECIMAL(4,1),     -- Consumo en ciudad (L/100 km)
    FUELCONSUMPTION_HWY DECIMAL(4,1),      -- Consumo en carretera (L/100 km)
    FUELCONSUMPTION_COMB DECIMAL(4,1),     -- Consumo combinado (L/100 km)
    FUELCONSUMPTION_COMB_MPG INT,          -- Consumo combinado en Millas por Galón (mpg)
    CO2EMISSIONS INT                       -- Emisiones de CO2 (g/km)
);
```

### 3. Listado limpio y único de los fabricantes (marcas) de vehículos
Escribe una consulta que devuelva las marcas (MAKE) disponibles en el dataset sin que se repitan.

```python
SELECT DISTINCT MAKE 
FROM fuel_consumption 
ORDER BY MAKE ASC;
```

* **Explicación**: La palabra clave `DISTINCT` remueve cualquier duplicado de la columna `MAKE`. Se agrega `ORDER BY MAKE ASC` para ordenar las marcas alfabéticamente (por ejemplo, desde ACURA, ASTON MARTIN, AUDI hasta VOLVO), lo cual facilita la lectura del listado.
* **Resultado Esperado**: El dataset cuenta con exactamente **39 marcas únicas**.

### 4. Filtrar vehículos con tamaño de motor mayor a 3.0 litros
Filtra el dataset para mostrar únicamente los vehículos que tengan un tamaño de motor (ENGINE_SIZE) mayor a 3.0 litros para analizar el segmento de alto cilindraje.

```python
SELECT MAKE, MODEL, ENGINESIZE, CYLINDERS, CO2EMISSIONS 
FROM fuel_consumption 
WHERE ENGINESIZE > 3.0 
ORDER BY ENGINESIZE DESC;
```

* **Explicación**: Filtramos las filas donde `ENGINESIZE` sea superior a `3.0` usando `WHERE ENGINESIZE > 3.0`.
* **Resultado Esperado**: Hay **552 registros** que cumplen con esta condición (automóviles de alto cilindraje).

### 7. Encontrar vehículos con consumo de combustible en ciudad inusualmente bajo o alto
Escribe una consulta para encontrar los vehículos que registren un consumo de combustible en ciudad inusualmente bajo o alto, aplicando filtros específicos en las columnas de consumo.

```python
SELECT MAKE, MODEL, VEHICLECLASS, FUELCONSUMPTION_CITY, CO2EMISSIONS
FROM fuel_consumption
WHERE FUELCONSUMPTION_CITY <= 6.0 
   OR FUELCONSUMPTION_CITY >= 25.0
ORDER BY FUELCONSUMPTION_CITY ASC;
```

* **Explicación**:
  - Un consumo de combustible en ciudad inusualmente bajo (alta eficiencia) se define aquí como menor o igual a **6.0 L/100 km** (típico de vehículos híbridos como el *HONDA Accord Hybrid* con 4.7 o *FORD Fusion Hybrid* con 5.3).
  - Un consumo inusualmente alto (baja eficiencia) se define como mayor o igual a **25.0 L/100 km** (típico de vehículos pesados de carga o transporte como la van *CHEVROLET Express 3500 Passenger* con 30.2).
  - La condición `OR` permite capturar ambos extremos de la distribución en una sola consulta.
* **Resultado Esperado**: Hay **26 vehículos** en total que caen en estos extremos (12 muy bajos y 14 muy altos).
